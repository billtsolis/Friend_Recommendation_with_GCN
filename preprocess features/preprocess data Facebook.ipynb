```python
import pandas as pd
import numpy as np
import os
import collections
```

```python
folder = "fb/facebook"
apath="snap_dataset/"
```

```python
def process_file(file_id, folder):
    # Read ego, egofeat, and featnames data from the file
    ego = pd.read_csv(apath+f"{folder}/{file_id}.egofeat", sep=" ", header=None)
    egofeat = pd.read_csv(apath+f"{folder}/{file_id}.feat", sep=" ", header=None)
    featnames = pd.read_csv(apath+f"{folder}/{file_id}.featnames", sep=" ", header=None)
    print(f"Processing file with ID: {file_id}")
    # Create a list of ego features
    eglist=[file_id] #set 1th element as file id
    for i in list(ego):
        eglist.append(int(ego[i].iloc[0])) #features list from ego
    egofeat.loc[len(egofeat.index)] = eglist #combine .egofeat with .feat
    
    #In case of the bad featurename split from csv edit the column
    featnames['feat_name'] = featnames.apply(lambda row: f"{row[0],row[1],row[2]}", axis=1)
    featnames = featnames.drop([1,2,3], axis=1)
    featnames[0] += 1 #reallign columns
    featnames.loc[len(egofeat.index)] = [0, 'id'] #set 0 as id
    featnames = featnames.sort_values(by=[0]) #sort values
    
    a = list(featnames['feat_name'])
    egofeat.columns = a #change the column names to match feature name 
     
    return egofeat

file_ids = []
for filename in os.listdir(apath+f"{folder}"):
    if filename.endswith(".egofeat"):
        file_id = int(filename.split(".")[0])
        file_ids.append(file_id)  
        
egofeat_list = [process_file(file_id, folder) for file_id in file_ids]
```

```python
result = egofeat_list[0]
for egofeat in egofeat_list[1:]:
    common_cols = set(result.columns) & set(egofeat.columns) # Find the common columns
    result = pd.merge(result, egofeat, on=list(common_cols), how='outer') # Merge ego feat and features
```

```python
result.sort_values(by=['id'])
```

```python
result2=result.replace(np.nan,0)

df_boolean = result2.iloc[:, 1:].astype(bool)

# Concatenate the first column with the boolean DataFrame
result2 = pd.concat([result2.iloc[:, :1], df_boolean], axis=1)
result3 = result2.groupby('id').max().reset_index()
result3=result3.set_index('id')
result3
```

```python
result3.to_csv(f"raw/fb.txt", sep=' ', index=False, header=False)
```
